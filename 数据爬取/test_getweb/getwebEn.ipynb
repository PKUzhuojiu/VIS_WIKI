本程序用于获取编辑历史条目，适用于英文维基百科

由于访问英文维基百科需要代理，请把proxies修改成自己的代理

如果报错Cannot connect to proxy，可以参考https://blog.csdn.net/qq_34151419/article/details/102670166

```python
# 获取编辑历史条目，适用于英文维基
# 输出见edit_history.csv
import requests as rq
from bs4 import BeautifulSoup
import csv
import unicodedata
import numpy

url = "https://en.wikipedia.org/w/index.php?title=Donald_Trump&action=history" # 修改成要爬取的词条的链接

old_ids = []
next_link = url
pageid = 1
cnt_oldpage = 0

f = open('edit_history.csv', 'w', encoding='utf-8', newline='')
f_write = csv.writer(f)

s = rq.session()
s.proxies = {"https": "127.0.0.1:1080"} # 修改成自己的代理
s.keep_alive = False

while (next_link):
    print(pageid, '...')
    pageid += 1
    r = s.get(next_link, timeout=60)
    print('get')
    # r = s.get(url)
    r.encoding = 'utf-8'
    soup = BeautifulSoup(r.text, 'lxml')
    for item in soup.find_all('li', {'data-mw-revid':True}):
        old_id = str(item.attrs['data-mw-revid'])
        cnt_oldpage += 1
        hist_info = [cnt_oldpage, old_id]

        def hist_append(obj):
            if obj:
                hist_info.append(obj.get_text())
            else:
                hist_info.append('')

        # date mw-changeslist-date
        date = item.find('a', class_='mw-changeslist-date')
        hist_append(date)

        # user userlink
        user = item.find('a', class_ = 'mw-userlink')
        hist_append(user)

        # totalsize
        totalsize = item.find('span', {'data-mw-bytes':True})
        # hist_append(totalsize)
        hist_info.append(totalsize.attrs['data-mw-bytes'])

        # diff-size
        diffsize = item.find('span', {'data-mw-bytes':False, 'class': 'mw-diff-bytes'})
        hist_append(diffsize)

        # commment
        comment = item.find('span', {'class':'comment'})
        hist_append(comment)

        def check_revert(comment) :
            flag1 = comment.find('Revert')
            flag2 = comment.find('revert')
            flag3 = comment.find('Undo')
            flag4 = comment.find('Undid')
            if (flag1 == -1 & flag2 == -1 & flag3 == -1 & flag4 == -1):
                return 0
            return 1

        # print(hist_info)
        if (comment):
            if (check_revert(comment.get_text())) :
                f_write.writerow(hist_info)
    nextLink = soup.find('a', class_='mw-nextlink')
    if nextLink:
        next_link = 'https://en.wikipedia.org' + str(nextLink.get('href'))
    else:
        break
        

f.close()

```
